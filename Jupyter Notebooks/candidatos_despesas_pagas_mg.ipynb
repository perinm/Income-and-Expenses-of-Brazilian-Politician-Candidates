Importando biblioteca pandas e configurando limite de colunas/linhas para possibilitar ampla exploração facilmente.

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

Lendo csv de despesas a fim de exploração variáveis do mesmo, e separar colunas de interesse para posterior análise.

```python
df = pd.read_csv('despesas_pagas_candidatos_2018_MG.csv', sep=';',
          encoding='latin1', decimal=',', parse_dates=[0])
df.head().T
```

```python
df_candidato = (df.groupby(['ST_TURNO','SQ_PRESTADOR_CONTAS'])
                .agg({'SQ_PARCELAMENTO_DESPESA':'count', 'VR_PAGTO_DESPESA': 'sum'})
                .reset_index()
                .rename(columns={"SQ_PARCELAMENTO_DESPESA":"Count_SQ_DESPESA",
                                 "VR_PAGTO_DESPESA":"GASTOS_TOTAIS"})
                )
```

```python
df_candidato.sort_values("GASTOS_TOTAIS",ascending=0)
```

```python
df = df[["SQ_PRESTADOR_CONTAS","DS_TIPO_DOCUMENTO","NR_DOCUMENTO","DS_FONTE_DESPESA","DS_ORIGEM_DESPESA","SQ_DESPESA","SQ_PARCELAMENTO_DESPESA","DS_DESPESA","VR_PAGTO_DESPESA"]]
```

```python
df1 = pd.read_csv('despesas_contratadas_candidatos_2018_MG.csv',
                 sep=';', encoding='latin1', decimal=',', parse_dates=[0])

df1.head().T
```

```python
df_candidato1 = (df1.groupby(['ST_TURNO','SQ_PRESTADOR_CONTAS', 'NR_CANDIDATO',
                "SQ_CANDIDATO",'NM_CANDIDATO','NR_PARTIDO','SG_PARTIDO'])
                .agg({'SQ_DESPESA':'count', 'VR_DESPESA_CONTRATADA': 'sum'})
                .reset_index()
                .rename(columns={"SQ_DESPESA":"Count_SQ_DESPESA",
                                "VR_DESPESA_CONTRATADA":"GASTOS_TOTAIS"})
                )
```

```python
df_candidato1.T
```

```python
pd.concat([df_candidato, df_candidato1]).groupby(['SQ_PRESTADOR_CONTAS']).sum().reset_index()
```
