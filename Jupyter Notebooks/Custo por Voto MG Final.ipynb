```python
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

Abrindo arquivos onde estão os votos somados

```python
df = pd.read_csv('Votos_Somados.csv', decimal=',')
df.head().T
```

Abrindo arquivo onde todas as despesas estão consolidadas

```python
df1 = pd.read_csv('Despesas_Consolidadas.csv',decimal=',')
df1.head().T
```

Filtrando arquivos para o primeiro turno

```python
df_f = df[df['NR_TURNO'] == 1]
df1_f = df1[df1['ST_TURNO'] == 1]
```

Executando merge nas colunas em comum

```python
merged = pd.merge(df_f,df1_f,on=["SQ_CANDIDATO",'SG_PARTIDO',"NM_CANDIDATO"])
```

```python
merged.T
```

Checando tipo de cada coluna para posterioes operações

```python
merged.info()
```

Acima foi visto que a coluna QT_VOTOS_NOMINAIS já está do tipo int<br />
Transformando a coluna Gastos totais em float para possibiltiar opeção de cáclulo do Custo por Voto

```python
merged = merged.astype({"GASTOS_TOTAIS": float})
merged["Custo_por_Voto"] = merged["GASTOS_TOTAIS"]/merged["QT_VOTOS_NOMINAIS"]
```

```python
merged.GASTOS_TOTAIS = merged.GASTOS_TOTAIS.astype("float64")
```

```python
merged.info()
```

```python
merged.T
```

Achando valores unicos de DS_DETALHE_SITUACAO_CAND e DS_SIT_TOT_TURNO

```python
merged["DS_SIT_TOT_TURNO"].unique()
```

```python
merged['DS_DETALHE_SITUACAO_CAND'].unique()
```

```python
Eleitos = merged[merged['DS_SIT_TOT_TURNO'].isin(['ELEITO POR QP','ELEITO POR MÉDIA','ELEITO'])]
Eleitos["QT_VOTOS_NOMINAIS"].astype('int64').describe()
```

```python
Eleitos["Custo_por_Voto"].describe()
```

Como é possível ver, o custo por voto médio por candidato eleito é de 10 reais arredondando <br />
Com o maior valor sendo 34 e o menor dando cerca de 34 centavos

Filtrando Custo por Voto sendo maior que o maior valor de de candidato eleito

```python
Laranjas = merged[merged.Custo_por_Voto>34.7]
```

Removendo casos onde a candidatura do candidato foi recusada por algum motivo x <br />
Filtrando por quantidade de votos sendo menor que 1000

```python
Laranjas = Laranjas[~Laranjas['DS_DETALHE_SITUACAO_CAND'].isin(['INDEFERIDO COM RECURSO','INDEFERIDO'])]
Laranjas = Laranjas[Laranjas.QT_VOTOS_NOMINAIS<1000]
```

Abaixo temos o resultado dessa análise, uma lista com possíveis candidatos a se investigar o paradeiro do dinheiro.

```python
Laranjas[["SG_PARTIDO","SQ_PRESTADOR_CONTAS","SQ_CANDIDATO","NM_CANDIDATO","DS_DETALHE_SITUACAO_CAND","DS_SIT_TOT_TURNO","QT_VOTOS_NOMINAIS","GASTOS_TOTAIS","Custo_por_Voto"]].sort_values(by=["GASTOS_TOTAIS"],ascending=0)
```
