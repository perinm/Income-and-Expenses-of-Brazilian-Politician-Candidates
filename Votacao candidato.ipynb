```python
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('votacao_candidato_munzona_2018_MG.csv', sep=';',
          encoding='latin1', decimal=',', parse_dates=[0],error_bad_lines=False)
df.head().T
```

```python
df = df[["NR_TURNO","SQ_CANDIDATO","NM_CANDIDATO",'DS_DETALHE_SITUACAO_CAND',"SG_PARTIDO","NM_COLIGACAO","DS_COMPOSICAO_COLIGACAO","DS_SIT_TOT_TURNO","QT_VOTOS_NOMINAIS"]]
```

```python
votos_por_cand_e_turno = df.groupby(['NR_TURNO','SG_PARTIDO','SQ_CANDIDATO','NM_CANDIDATO','DS_DETALHE_SITUACAO_CAND','DS_SIT_TOT_TURNO'])['QT_VOTOS_NOMINAIS'].agg('sum').reset_index()
```
